## 모분산 가설 검정

### 모 평균을 아는 경우 (자유도 n인 카이제곱분포를 따름)

T = n * S^2 / sigma0^2

S^2 = 1/n * ((X1 - u)^2 + (X2 - u)^2 + ... + (Xn - u)^2)

T는 자유도 n인 카이제곱분포를 따름

### 모 평균을 모르는 경우 (자유도 n-1인 카이제곱분포를 따름)

T = (n-1) * U^2 / sigma0^2

U^2 = 표본 분산

T는 자유도 n-1인 카이제곱분포를 따름

- data = [172, 175, 178, 182, 176, 180, 169, 185, 173, 177]  
- 모평균을 모를때
- 모분산이 5^2 인지 아닌지 유의수준 5%로 가설검정

```python
# H0: sigma^2 = 5^2
# H1: sigma^2 != 5^2 (양측검정)
```

```python
from scipy.stats import t, chi2

data = [172, 175, 178, 182, 176, 180, 169, 185, 173, 177]
n = len(data)
s = np.std(data, ddof=1)
alpha = 0.05
sigma0 = 5

chi2_value = (n-1)*(s**2) / (sigma0**2)
p_value = chi2.cdf(chi2_value, n-1) *2

#양측 기각역 chi2 통계량 계산
chi2_alt_left = chi2.ppf(alpha/2, n-1)
chi2_alt_right = chi2.ppf(1-alpha/2, n-1)
```

```python
p_value, chi2_value, chi2_alt_left, chi2_alt_right
```

```python
chi2_value
```

```python
chi2.cdf(chi2_value, n-1)
```

```python
# chi2 통계량이 기각역에 있지 않음, chi2_alt_left < chi2_value < chi2_alt_right
# 귀무가설을 기각하지 못한다.
# H0: sigma^2 = 5^2
```

## alpha와 beta, 검정력

|           | H0 기각           | H0 수용          |
| --------- | ----------------- | ---------------- |
| H0가 참   | 제1종오류 (alpha) | O                |
| H1가 참 | O                 | 제2종오류 (beta) |



### 1번문제

```python
# H0: u=1500
# H1: u=1600 (복합가설)
# alpha = H0가 참인데, H0을 기각할 확률 (위험도)
# beta = H1이 참인데, H0를 수용할 확률 (검정력)
# 검정력을 구할때는, alpha에 해당하는 기각역을 구한뒤 beta의 면적을 구한다.

# H0 표본평균은 N(1500,200^2/30)의 분포를 따른다.
# H1 표본평균은 N(1600,200^2/30)의 분포를 따른다.
```

```python
# 유의수준 5%의 기각역 계산

sigma = 200
n = 30
alpha = 0.05
u0 = 1500
u1 = 1600

# [X, X+z0.95*sigma/n**0.5]
z095 = norm.ppf(1-alpha)
u_alt = u0 + (z095 * sigma/n**0.5)
u_alt
```

```python
# 유의수준 5%의 기각역에 해당하는 beta 구하기
# H1이 참인데도 불구하고, H1이 기각될 확률

z1 = (1600-u_alt)/(sigma/n**0.5)
z1 #H1의 기각역
```

```python
# beta = z1이 기각역의 면적
beta = 1-norm.cdf(z1)
beta
```

```python
# 검정력
1-beta
```
