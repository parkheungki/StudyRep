### 통계 문제 연습 

scipy.stats 의 binom, norm 등의 툴을 이용해서 풀어보세요.  

```python
from scipy.stats import binom #scipy.stat 패키지 import
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1,1) #subplot 만들기
n = 16 #총 시도횟수
p = 0.7 #각 독립사건 성공확률
x = np.arange(binom.ppf(q = 0, n = n, p = p), binom.ppf(q = 1, n = n, p = p))
'''
ppf(q, n, p) : percent point function
            -> # of saample = n, prob = p 일때, q이상인 k를 구함
            참고 : https://stackoverflow.com/questions/60699836/how-to-use-norm-ppf
'''
ax.plot(x, binom.pmf(x, n, p),'bo', ms=8, label = 'binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw =5, alpha =0.5)
```

#### (b) 10명 미만의 승객이 짐을 수화물로 보낼 확률

`scipy.stats.binom.cdf(k, n, p)` : 	Cumulative distribution function

- k : 성공횟수
- n : 총 횟수
- p : 사건확률
- k, n, p의 누적 이산확률 질량함수의 확률값 계산

```python
from scipy.stats import binom #scipy.stat 패키지 import
import matplotlib.pyplot as plt
import numpy as np
n = 16 #총 시도횟수
p = 0.7 #각 독립사건 성공확률
k = 9 #성공 횟수
print(binom.cdf(k = k, n = n, p = p))
```

```python
from scipy.stats import binom #scipy.stat 패키지 import
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1,1) #subplot 만들기
n = 16 #총 시도횟수
p = 0.7 #각 독립사건 성공확률
x = np.arange(binom.ppf(q = 0, n = n, p = p), binom.ppf(q = 1, n = n, p = p))
'''
ppf(q, n, p) : percent point function
            -> # of saample = n, prob = p 일때, q이상인 k를 구함
            참고 : https://stackoverflow.com/questions/60699836/how-to-use-norm-ppf
'''
ax.plot(x, binom.cdf(x, n, p),'bo', ms=8, label = 'binom pmf')
ax.vlines(x, 0, binom.cdf(x, n, p), colors='b', lw =5, alpha =0.5)
```

```python
s = 0
for i in range(17):
    x = binom.pmf(k=i, n=16, p=0.7)
    s += x
    print(f'{i}번째 : x = {x}, sum = {s}')
```

#### (c) 16명의 승객을 무작위로 추출하여 그 중 짐을 수화물로 보내는 승객의 수를 세는 실험을 1000번 하는 상황을 시뮬레이션 하여 그 데이터의 히스토그램을 그리시오.

`rvs(n, p, loc=0, size=1, random_state=None)` : random variable sampling

- n : 이산확률분포 총 횟수
- p : 사건 확률

```python
n = 16
p = 0.7
r = binom.rvs(n, p, size=1000)
```

```python
import collections
import matplotlib.pyplot as plt
counter = collections.Counter(r)
plt.bar(counter.keys(), counter.values())
```

#### (d) 위의 시뮬레이션 데이터의 평균과 분산은 무엇인가? 이는 이론적인 기대값, 분산과 일치하는가?

```python
m = np.mean(r)
s = np.std(r)
print(f'mean = {m}, std = {s}')
```

```python
import math
n = 16
p = 0.7
theory_m = n * p
theory_std = math.sqrt(n * p * (1 - p))

print(f'이론적인 이산확률분포 평균 : {theory_m}')
print(f'이론적인 이산확률분포 분산 : {theory_std}')
```

   - 가설
        - H1: 도전자가 선택한 Room(A라 한다)에 상금 존재
        - H2: Open된 Room(B라 한다)에 상금 존재
        - H3: 남은 Room(C라 한다)에 상금 존재  
          
   - 관측 데이터 
     
        - 사회자가 Room B를 Open   
        
        
   - 가설 1,2,3에 대한 사전분포는 동일하게 1/3로 가정한다. 즉, P(H1)=P(H2)=P(H3)=1/3이다.

##### (2) 위의 (1)과 비슷하게 P(D|H2), P(D|H3)을 구하시오.

##### (3) 사회자가 Room B를 Open 했을 때 상금이 A에 존재할 확률, P(H1|D),과 상금이 C에 존재할 확률, P(H3|D),을 계산하시오. 계산 결과를 바탕으로 도전자에게 유리할 의사결정이 무엇인지 답하시오.`
